```python
import pandas as pd

df = pd.read_table('34658-0001-Data.tsv', sep='\t', low_memory=False)
print(df.shape)
df.head()
```

```python
cols_of_intrest = ['NHOOD02','NHOOD04','NHOOD10','NHOOD12',    # Good neighbourhood features
                   'NHOOD05','NHOOD06','NHOOD07','NHOOD08','NHOOD09',  # Bad neighbourhood features
                   'BEERLTP','SPIRLTP', # ALCOHOL CONSUMPTION                
                   'HASHLTP','XTCLTP','LHCLTP', # DRUG CONSUMPTION                  
                   'VANDLTP','SHOPLTP','BURGLTP','BICTLTP','CARTLTP','CARBLTP','SNATLTP', # THEFT ACTIVITIES      
                   'WEAPLTP','EXTOLTP','GFIGLTP','ASLTLTP', # VIOLENCE                   
                   'DRUDLTP' # DRUG SELLING"
                    
                  ]
                   
df_neigh = df.filter(items = cols_of_intrest)
df_neigh.columns
```

```python
def neigh(row):
    if row['NHOOD05'] in [3, 4] or row['NHOOD06'] in [3, 4] or row['NHOOD07'] in [3, 4] or row['NHOOD08'] in [3, 4] or row['NHOOD09'] in [3, 4]:
        return 'Bad'
    elif row['NHOOD02'] in [3, 4] or row['NHOOD04'] in [3, 4] or row['NHOOD10'] in [3, 4] or row['NHOOD12'] in [3, 4]:
        return 'Good'
    elif row['NHOOD02'] in [1, 2] and row['NHOOD04'] in [1, 2] and row['NHOOD10'] in [1, 2] and row['NHOOD12'] in [1, 2]:
        return 'Bad'
    elif row['NHOOD05'] in [1, 2] and row['NHOOD06'] in [1, 2] and row['NHOOD07'] in [1, 2] and row['NHOOD08'] in [1, 2] and row['NHOOD09'] in [1, 2]:
        return 'Good'
    else:
        return 'No Answer'
    

def alc_con(row):
    if row['BEERLTP'] == 1 or row['SPIRLTP'] == 1:
        return 'Yes'
    elif row['BEERLTP'] == 0 and row['SPIRLTP'] == 0:
        return 'No'
    elif row['BEERLTP'] == 7 or row['SPIRLTP'] == 7:
        return 'Ambiguous'
    else:
        return 'Not Applicable'


def drug_con(row):
    if row['HASHLTP'] == 1 or row['XTCLTP'] == 1 or row['LHCLTP'] == 1:
        return 'Yes'
    elif row['HASHLTP'] == 0 and row['XTCLTP'] == 0 and row['LHCLTP'] == 0:
        return 'No'
    elif row['HASHLTP'] == 7 or row['XTCLTP'] == 7 or row['LHCLTP'] == 7:
        return 'Ambiguous Answer'
    else:
        return 'No Answer'

    
def theft_act(row):
    if row['VANDLTP'] == 1 or row['SHOPLTP'] == 1 or row['BURGLTP'] == 1 or row['BICTLTP'] == 1 or row['CARTLTP'] == 1 or row['CARBLTP'] == 1 or row['SNATLTP'] == 1:
        return 'Yes'

    elif row['VANDLTP'] == 0 and row['SHOPLTP'] == 0 and row['BURGLTP'] == 0 and row['BICTLTP'] == 0 and row['CARTLTP'] == 0 and row['CARBLTP'] == 0 and row['SNATLTP'] == 0:
        return 'No'
    
    elif row['VANDLTP'] == 7 or row['SHOPLTP'] == 7 or row['BURGLTP'] == 7 or row['BICTLTP'] == 7 or row['CARTLTP'] == 7 or row['CARBLTP'] == 7 or row['SNATLTP'] == 7:
        return 'Ambiguous'
    else:
        return 'No Answer'

def vio_act(row):
    if row['WEAPLTP'] == 1 or row['EXTOLTP'] == 1 or row['GFIGLTP'] == 1 or row['ASLTLTP'] == 1:
        return 'Yes'
    elif row['WEAPLTP'] == 0 and row['EXTOLTP'] == 0 and row['GFIGLTP'] == 0 and row['ASLTLTP'] == 0:
        return 'No'
    elif row['WEAPLTP'] == 7 or row['EXTOLTP'] == 7 or row['GFIGLTP'] == 7 or row['ASLTLTP'] == 7:
        return 'Ambiguous'
    else:
        return 'No Answer'


def drug_sell(row):
    if row['DRUDLTP'] == 1:
        return 'Yes'
    elif row['DRUDLTP'] == 0:
        return 'No'
    elif row['DRUDLTP'] == 7:
        return 'Ambiguous'
    else:
        return 'No Answer'
```

```python
df_neigh['Neighborhood'] = df_neigh.apply(neigh, axis=1)
df_neigh['Alcohol_Consumption'] = df_neigh.apply(alc_con, axis=1)
df_neigh['Drug_Consumption'] = df_neigh.apply(drug_con, axis=1)
df_neigh['Theft_Activity'] = df_neigh.apply(theft_act, axis=1)
df_neigh['Violence_Activity'] = df_neigh.apply(vio_act, axis=1)
df_neigh['Drug_Selling'] = df_neigh.apply(drug_sell, axis=1)
fin_cols = ['Neighborhood', 'Alcohol_Consumption', 'Drug_Consumption', 'Theft_Activity', 'Violence_Activity', 'Drug_Selling']
df_final = df_neigh[fin_cols]
df_final.to_csv('aggregated.csv')
df_final.head()
```
